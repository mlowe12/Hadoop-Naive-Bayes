```python
row, col = table.shape
arr = []
for i in range(row):
    temp = str(table[headers[0]][i])
    print(temp)
    #regex out all special characters from each text message before tokenizing each word. 
    arr.append((re.search("[^&^!@#$%^&*()_\-+=|/\\<>?\'\";:\[\]{}`~.]+", str(temp)).group(0)))
    
    
    
```

```python
def produceWordBag(list):
    temp_arr = []
    for i in range(len(list)):
        e = re.search("\w+",list[i]).group(0)
        temp_arr.append(str(e))
    return temp_arr
```
